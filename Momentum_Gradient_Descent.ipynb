```python
############
#
# Momentum Gradient Descent
# (Deep Learning Training Example)
#
############

### Import libraries
import numpy as np
import matplotlib.pyplot as plt

### Create dataset
np.random.seed(42)

X = np.linspace(0,10,100)
y = 3*X + 5 + np.random.randn(100)*2

X = X.reshape(-1,1)
y = y.reshape(-1,1)

### Hyperparameters
learning_rate = 0.001
epochs = 50
momentum = 0.9

### Initialize weights randomly
weights = np.random.randn(1,1)
bias = np.random.randn(1)

# velocity terms (momentum)
v_w = np.zeros((1,1))
v_b = np.zeros(1)

print("Initial Weight:", weights[0][0])
print("Initial Bias:", bias[0])

### Store loss values
loss_history = []

### Training using Momentum Gradient Descent
for epoch in range(epochs):

    # prediction
    y_pred = np.dot(X, weights) + bias

    # error
    error = y - y_pred

    # loss (MSE)
    loss = np.mean(error**2)
    loss_history.append(loss)

    # gradients
    dW = -2 * np.dot(X.T, error) / len(X)
    dB = -2 * np.mean(error)

    # momentum update
    v_w = momentum * v_w + learning_rate * dW
    v_b = momentum * v_b + learning_rate * dB

    # update parameters
    weights = weights - v_w
    bias = bias - v_b

    print(f"Epoch {epoch+1:3d} | Loss: {loss:.4f}")

print("\nTraining Complete")
print("Final Weight:", weights[0][0])
print("Final Bias:", bias[0])

### Predictions
y_pred_all = np.dot(X, weights) + bias

### Plot regression result
plt.figure(figsize=(8,5))

plt.scatter(X, y, color='blue', label="Data Points")
plt.plot(X, y_pred_all, color='red', linewidth=2, label="Momentum GD Line")

plt.xlabel("X values")
plt.ylabel("Target values")
plt.title("Momentum Gradient Descent Regression")
plt.legend()
plt.grid(True)

plt.show()

### Plot loss curve
plt.figure(figsize=(6,4))
plt.plot(loss_history, color='green')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss Curve (Momentum GD)")
plt.grid(True)
plt.show()
```
